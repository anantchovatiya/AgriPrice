```python
import gradio as gr
import numpy as np
from datetime import datetime, timedelta

# Mock data for states, districts, markets, and commodities
state_district_map = datacopy.groupby('State')['District'].unique().apply(list).to_dict()
district_market_map = datacopy.groupby('District')['Market'].unique().apply(list).to_dict()
commodity_market_map = datacopy.groupby('Market')['Commodity'].unique().apply(list).to_dict()

# Function to dynamically update districts based on selected state
def update_districts(state):
    districts = state_district_map.get(state, [])
    return gr.update(choices=districts, value=districts[0] if districts else None)

# Function to dynamically update markets based on selected district
def update_markets(district):
    markets = district_market_map.get(district, [])
    return gr.update(choices=markets, value=markets[0] if markets else None)

# Function to dynamically update commodities based on selected market
def update_commodities(market):
    commodities = commodity_market_map.get(market, [])
    return gr.update(choices=commodities, value=commodities[0] if commodities else None)

# Main function to predict price
def predict_price_gradio(commodity, state, district, market, date):
    commodity_value, state_value, district_value, market_value, month_value, year_value, day_value = map_user_input(
        commodity, state, district, market, date
    )

    predicted_price = round(predict_price(commodity_value, state_value, district_value, market_value, day_value, month_value, year_value), 2)
    uinput = [list(dictOfWord_comm.keys())[commodity_value], list(dictOfWord_state.keys())[state_value], list(dictOfWord_dis.keys())[district_value], list(dictOfWord_mar.keys())[market_value]]
    yesterday_price = round(yes_price(uinput), 2)
    yesterday = datetime.now() - timedelta(days=1)
    date_value = yesterday.strftime('%d/%m/%Y')

    if predicted_price is not None:
        return f"""
        <div style='border: 2px solid #ccc; padding: 20px; border-radius: 10px;'>
            <p style='font-size: 22px; text-align: center;'><strong>Predicted Modal Price:</strong> {predicted_price} per 100KG</p>
            <p style='font-size: 28px; color: #ff5733; text-align: center; animation: flicker 1.5s infinite;'><strong>Yesterday's Price:</strong> {yesterday_price} per 100KG</p>
        </div>
        """
    else:
        return "Invalid input: Please ensure all inputs are within the dataset."

# Custom CSS for flicker effect and website style
css = """
    @keyframes flicker {
        0% { opacity: 1; }
        50% { opacity: 0.3; }
        100% { opacity: 1; }
    }
    .title {
        font-size: 40px;
        text-align: center;
        font-weight: bold;
        margin-top: 20px;
        color: #2c3e50;
    }
    .heading {
        font-size: 32px;
        text-align: center;
        margin-top: 10px;
        color: #16a085;
    }
    .gradio-container {
        max-width: 800px;
        margin: 0 auto;
        padding: 20px;
        background-color: white;
        box-shadow: 0px 0px 15px rgba(0, 0, 0, 0.1);
        border-radius: 10px;
    }
    body {
        background-color: #ecf0f1;
    }
"""

# Homepage content
def homepage():
    return """
    <div style='text-align: center; padding: 20px;'>
        <h1 style='color: #2c3e50;'>Welcome to AGRIPRICE</h1>
        <p style='font-size: 20px;'>AGRIPRICE is a powerful tool for predicting agricultural prices across various commodities and markets in India.</p>
        <p style='font-size: 16px;'>Explore our platform to find daily predictions and insights on market trends.</p>
        <p> Developed by Anant Chovatiya , Jasmin Chaudhary </p>
    </div>
    """

# Contact content
def contact():
    return """
    <div style='text-align: center; padding: 20px;'>
        <h2>Contact Us</h2>
        <p>If you have any questions or need support, reach out to our team:</p>
        <p>Email: support@agriprice.com</p>
        <p>Phone: +91 123-456-7890</p>
    </div>
    """

# Create Gradio interface with enhanced UI
with gr.Blocks(css=css) as iface:
    with gr.Tabs():
        # Home page
        with gr.TabItem("Home"):
            gr.HTML(homepage)

        # Prediction page
        with gr.TabItem("Predict Price"):
            gr.Markdown("<div class='title'>AGRIPRICE</div>")

            # Input fields
            state_input = gr.Dropdown(list(state_district_map.keys()), label="State")
            district_input = gr.Dropdown([], label="District")
            market_input = gr.Dropdown([], label="Market")
            commodity_input = gr.Dropdown([], label="Commodity")
            date_input = gr.Textbox(label="Date (dd-MM-YYYY)")

            # Link state selection to update districts
            state_input.change(fn=update_districts, inputs=state_input, outputs=district_input)

            # Link district selection to update markets
            district_input.change(fn=update_markets, inputs=district_input, outputs=market_input)

            # Link market selection to update commodities
            market_input.change(fn=update_commodities, inputs=market_input, outputs=commodity_input)

            # Button for price prediction
            predict_button = gr.Button("Predict Price")
            output = gr.HTML(label="Predicted Price and Yesterday's Price")

            # Predict the price on button click
            predict_button.click(
                fn=predict_price_gradio,
                inputs=[commodity_input, state_input, district_input, market_input, date_input],
                outputs=output
            )

        # Contact page
        with gr.TabItem("Contact"):
            gr.HTML(contact)

iface.launch(share=True)
```
